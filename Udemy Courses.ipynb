### **Univariate Analysis**

```python
bin_edges = [0, 20,  40,  60,  80]
pt=sns.histplot(data=df, x='price', bins=bin_edges,  color='#eb44e8',edgecolor='#54504c')
plt.xlabel('Course Prices',color='#2e2b28', fontweight='bold')
plt.ylabel('Count',color='#2e2b28', fontweight='bold')
plt.title("Distribution of course prices", fontweight="bold", color='#6b506b')
plt.xticks(color='#2e2b28')
plt.yticks(color='#2e2b28')
for i in pt.containers:
    pt.bar_label(i, padding=-20, color='white', fontweight='bold')
```

#### Observations

- Approximately 48% (1262 courses) of the courses are priced between 20-40$, making it the most common price range.

```python
plt.figure(figsize=(8,4))
pt=sns.countplot(x='subject', data=df, order=df['subject'], color='#eb44e8', edgecolor='#54504c')
plt.title("Number of Courses per Subject", fontweight="bold", color='#6b506b')
plt.xlabel('Subject',fontweight='bold', color='#2e2b28')
plt.ylabel('Count of Courses', fontweight='bold', color='#2e2b28')
plt.xticks(color='#2e2b28')
plt.yticks(color='#2e2b28')
for i in pt.containers:
    pt.bar_label(i, padding=-20, color='white', fontweight='bold')
```

#### Observations

-  Web Development is the most popular subject with 1,199 courses, closely followed by Business Finance with 1,191 courses.
- Subjects like Musical Instruments and Graphic Design have a smaller but notable presence.

# Multivariate Analysis

```python
custom_palette = sns.color_palette(['#fec0ff','#fc93e8', '#ee4fc5', '#ab3da9'])
sns.scatterplot(data=df, x='price', y='num_subscribers', hue='subject', palette='viridis')
plt.title("Subscriber Distribution by Price and Subjects", fontweight="bold", color='#6b506b')
plt.xlabel('Course Price',fontweight='bold', color='#2e2b28')
plt.ylabel("Subscriber's Count", fontweight='bold', color='#2e2b28')
plt.xticks(color='#2e2b28')
plt.yticks(color='#2e2b28');
```

#### Observations

- Majority of courses with the highest subscriber counts are concentrated around free and low-priced courses (between 0-25$), particularly in Web Development.

```python
df_numerical = df.select_dtypes(include='number')
corr_matrix=df_numerical.corr()
sns.heatmap(data=corr_matrix, annot=True, cmap='Purples')
plt.title("Correlation Analysis of Numerical Variables", fontweight="bold", color='#6b506b')
plt.xticks(rotation=45,color='#2e2b28')
plt.yticks(color='#2e2b28');
```

#### Observations

- Content Duration & Number of lectures have high correlation with each other.
- Number of Subscribers has a good correlation with Number of reviews.

```python
df['Year']= df['published_timestamp'].dt.year
d=df.groupby('Year')['subject'].value_counts().reset_index()
pt=sns.lineplot(data=d, x=d['Year'], y=d['count'], hue=d['subject'], palette='viridis')
plt.legend(title='Subjects',bbox_to_anchor=(1,1))
plt.title("Count of Subject Courses by Year", fontweight="bold", color='#6b506b')
plt.xlabel('Year',fontweight='bold', color='#2e2b28')
plt.ylabel('Count of subject Courses', fontweight='bold', color='#2e2b28')
plt.xticks(color='#2e2b28')
plt.yticks(color='#2e2b28');
```

#### Observations

- The line chart reveals that Web Development reached its peak in 2016, followed by a decline in 2017, a similar trend is observed in Business Finance and Musical Instruments.
-  The dip suggests a shift in focus or interest during 2017, potentially due to market saturation or shifting demand toward other emerging subjects.
